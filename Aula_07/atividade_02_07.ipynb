```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('monitoramento_tempo.csv', sep=',')
```

```python
df.head()
```

```python
df['data']=pd.to_datetime(df['data'])
```

```python
df.info()
```

```python
%matplotlib inline
plt.plot(df['data'], df['humidade'])
```

```python
fig = plt.figure(figsize=(15, 8))

plt.rcParams['font.size']='20'
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['humidade'], color='#00FA9A', linestyle='-', marker='D')
eixo.set_title("humidade at the moment", fontsize=25)
eixo.legend("Humidade", loc='lower left', fontsize=20)
eixo.set_ylabel("humidade", fontsize= 20)
eixo.set_xlabel("Ano", fontsize= 20)
```
