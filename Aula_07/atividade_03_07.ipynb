```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('monitoramento_tempo.csv', sep=',')
```

```python
df.head()
```

```python
df['data']=pd.to_datetime(df['data'])
```

```python
df.info()
```

```python
%matplotlib inline
plt.plot(df['data'], df['direção do vento'])
```

```python
import datetime
```

```python
fig = plt.figure(figsize=(15, 8))

plt.rcParams['font.size']='15'
eixo = fig.add_axes([0,0,1,1])
#eixo.set_xlim(datetime.datetime(2014,8,1)) - período especifico
eixo.plot(df['data'], df['direção do vento'], lw=0.1, color='#DC143C', linestyle='-', marker='+')
eixo.set_title("direção do vento at the moment", fontsize=25)
eixo.legend("Direção do vento", loc='lower left', fontsize=20)
eixo.set_ylabel("direção do vento", fontsize= 16)
eixo.grid(True)
eixo.set_xlabel("Ano", fontsize= 16)
```

```python
df['temperatura'] = (df['temperatura']-32)*(5/9)
```

```python
fig = plt.figure(figsize=(15, 8))
plt.rcParams['font.size']='20'
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['data'], df['temperatura'], color = 'b', lw=0.5, linestyle='--', marker='x')
eixo.set_xlim(datetime.datetime(2014, 4, 1), datetime.datetime(2014,7,1))
eixo.set_title('Temperatura', fontsize=25)
eixo.legend(["Temperatura"], loc= 'lower right', fontsize=15)
eixo.set_xlabel('Ano', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.grid(True)
```

```python
fig = plt.figure(figsize=(15, 12))
plt.rcParams['font.size']='10'
eixo = fig.add_axes([0,0,1,1])
eixo2 = fig.add_axes([0.7,0.65,0.3,0.3])
eixo.set_xlim(datetime.datetime(2015, 4, 1), datetime.datetime(2015,7,1))
eixo.plot(df['data'], df['temperatura'], color = 'b', lw=0.5, linestyle='--', marker='x')

eixo.set_title("Temperatura", fontsize=25)
eixo.set_ylim(50, 100)
eixo.legend(["Temperatura"], loc= 'lower right', fontsize=15)
eixo.set_xlabel('Ano', fontsize=20)
eixo.set_ylabel('Temperatura', fontsize=20)
eixo.grid(True)
#eixo 2
eixo2.set_xlim(datetime.datetime(2015, 4, 1), datetime.datetime(2015, 8, 1))
eixo2.plot(df['data'], df['temperatura'], color='r')
eixo2.grid(True)
eixo2.set_title("Temperatura Abril - Julho", fontsize=10, pad=20)
eixo2.set_xlabel("Mês", fontsize=10)
eixo2.set_ylabel("Temperatura", fontsize=10)
eixo2.legend(['Temperatura'], loc='best', fontsize=10)

```
