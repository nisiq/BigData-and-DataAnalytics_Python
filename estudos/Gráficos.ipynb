```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
Dados_x = np.linspace(0, 10, 10)
Dados_y = Dados_x * Dados_x
```

```python
#dados excel
#ler arquivo excel:
df=pd.read_csv('dados.csv', sep=',')
```

```python
plt.title('Grfico eixos',
fontweight='bold',
fontsize='xx-large',
fontfamily='fantasy',
color='blue'
)
plt.plot(Dados_x, Dados_y,
color = 'red', linewidth = 2,
markersize = 5,
marker = 'o',
linestyle = 'dashed',
markeredgecolor = 'black',
)
plt.legend('numeros')
loc = 'low right',
bbox_to_anchor = (1, 0, 0.0, 0.2),
fontsize='large',
shadow = True,
borderpad=1,

plt.xlabel('Eixo X',
fontweight='bold',
fontsize= 'large',
fontfamily='fantasy',
color= 'blue'
)
plt.ylabel('Eixo Y',
fontweight='bold',
fontsize= 'large',
fontfamily='fantasy',
color= 'blue'
)
#plt.savefig('Grafico total ciot .eps', dpi=300, transparent=True, bbox_inches='tight')

#df['x1'].plot(kind='bar', color='b', marker='x')
```
